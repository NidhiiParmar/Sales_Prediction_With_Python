```python
df['TV']=df['TV'].astype('int')
df['Radio']=df['Radio'].astype('int')
df['Newspaper']=df['Newspaper'].astype('int')
df['Sales']=df['Sales'].astype('int')
df
```

```python
df.isnull().sum()
```

```python
x=df.drop('Sales',axis=1)
x
```

```python
x = x.drop(columns='Unnamed: 0')
x
```

```python
y=df[['Sales']]
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train,y_test = train_test_split(x,y,test_size = 0.20, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
# creating a Linear regression object
lr = LinearRegression()
lr
```

```python
lr.fit(x_train,y_train)
```

```python
coefficients = lr.coef_[0]
coefficients_df = pd.DataFrame({'Feature': ['TV', 'Newspaper', 'Radio'], 'Coefficient': coefficients})

print(coefficients_df)
```

```python
lr.intercept_
```

```python
y_pred=lr.predict(x_test)
```

```python
y_pred
```

```python
from sklearn import metrics
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
print('Mean Absolute Error: ',metrics.mean_absolute_error(y_pred,y_test))
```

```python
print('Root Mean Squared Error:',np.sqrt(metrics.mean_squared_error(y_pred,y_test)))
```

```python
print('R-Squared: ',metrics.r2_score(y_pred,y_test))
```
